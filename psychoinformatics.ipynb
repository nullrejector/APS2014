<h3>Loading the data</h3>

The Sexuality IAT data are provided in SPSS format on the OSF project page. But we don't like SPSS, because SPSS is proprietary and mean. We like tab-delimited files, which are open and friendly. So let's define a method to help us convert the SPSS files into something pandas can work with. We'll take advantage of the savReaderWriter package, which allows us to avoid SPSS entirely.

While we're at it, we'll also write a couple of methods to load and merge the data into one giant pandas DataFrame, which is what we'll operate on the rest of the way. We have data from 10 years (2004 - 2013), so we'll need to load each file in turn, and then merge them all. We'll also take the opportunity to rename some of the columns and values to something more intuitive.

```python
# More descriptive column names
COLUMN_MAP = {
    'D_biep.Straight_Good_all': 'IAT_overall',
    'raceomb': 'race',
    'ethnicityomb': 'ethnicity',
    'countrycit': 'citizenship',
    'countryres': 'residence',
    'Stimuli': 'stimuli',
    'sexuality': 'orientation'
}

def replace_values(data):
    ''' Replace numeric column values with more meaningful strings. '''
    
    cols = list(data.columns)
    if 'race' in cols:
        data['race'] = data['race'].replace({1: 'NatAmer', 2: 'EastAsian', 3: 'SouthAsian',
                                             4: 'PacIsl', 5: 'Black', 6: 'White', 7: 'Bl/Wh',
                                             8: '> 1 race', 9: 'Other'})
    if 'stimuli' in cols:
        data['stimuli'] = data['stimuli'].replace({1: 'GayMen', 2: 'Lesbians'})
    if 'orientation' in cols:
        data['orientation'] = data['orientation'].replace({1:'Hetero-', 2:'Homo-', 3:'Bi-' })
    return data

def sav_to_pandas(filename):
    """ Read in SPSS .sav file and convert to pandas DataFrame.
    Requires savReaderWriter. On OS X, installation requires additional 
    steps--see http://pythonhosted.org//savReaderWriter/. 
    Args:
        filename: path to .sav file on disk.
    Returns: a pandas DataFrame containing all the converted SPSS data.
    """
    sav = srw.SavReader(filename, ioLocale='en_US.UTF-8', verbose=True)
    rows = []
    for i in range(len(sav)):
        try:
            rows.append(sav[i])
        except: pass
    return pd.DataFrame(rows, columns=sav.varNames).rename(columns=COLUMN_MAP)

def load_data(years=None, save=False):
    """ Loads multiple years of data into a single df. Assumes a folder structure 
    like that found in the OSF project repository: one folder per year named 
    Sexuality_IAT_public_20??, with a .sav file below that.
    Args:
        years: optional list of years to load. If None, tries to load all years.
        save: when True, saves a tab-delimited version of the file to disk.
    """
    if years is None:
        years = range(2004, 2014)
    datasets = {}
    for y in years:
        print "Processing %d..." % y
        filename = '/Users/tal/data/iat-sexuality/Sexuality_IAT_public_%d/Sexuality IAT.public.%d.sav' % (y, y)
        try:
            data = sav_to_pandas(filename)
            datasets[y] = data
            if save:
                data.to_csv(filename.replace('.sav', '.csv'), index=False)
        except Exception, e:
            print e
    return datasets

def merge_datasets(datasets, intersect=False):
    """ Concatenate data from multiple years into a single DataFrame.
    Args:
        datasets: A dict with years as keys and pandas DataFrames as values.
        intersect: When True, only columns common to all years will be retained.
            When False (default), will return the union of columns, filling in 
            missing values with NaNs.
    Returns: a pandas DataFrame. """
    if intersect:
        shared_cols = list(set.intersection(*[set(d.columns) for d in datasets.values()]))
    for year, d in datasets.items():
        if intersect:
            d = d[shared_cols]
        d['year'] = year
        datasets[year] = d
    return pd.concat(datasets.values(), axis=0)
```

With the above methods defined, we can now get to the business of actually reading and loading the data.

```python
# Load all of the individual files into a list of DataFrames
datasets = load_data(save=False)
# Merge them into one big DF
data = merge_datasets(datasets)
# Replace values with sensible strings
data = replace_values(data)
```

```python
def plot_kde_by_factor(data, factors, min_cases=5000, y='IAT_overall', height=4, aspect=3):
    """ Plt a series of KDE plots displaying the density estimates for different levels of each factor. """
    
    fig, ax = plt.subplots(len(factors), 1, figsize=(height*aspect, height*len(factors)), sharey=True)
    
    if len(factors) == 1:
        ax = [ax]

    for fac_i, f in enumerate(factors):
        f_data = data[[f, y]].groupby(f).filter(lambda x: np.sum(~np.isnan(x[y])) >= min_cases)
        levels = f_data[f].unique()
        cols = sns.color_palette("hls", len(levels))
        for lev_i, l in enumerate(levels):
            level_data = f_data[y][f_data[f]==l]
            sns.kdeplot(level_data, shade=False, color=cols[lev_i], label=l, ax=ax[fac_i], lw=3)
            ax[fac_i].set_title(f, fontsize=18)
    fig.subplots_adjust(hspace=0.1)
    plt.tight_layout()
    

def plot_corr_matrix(data, factors, min_cases=5000, size=3):

    n_factors = len(factors)
    fig, ax = plt.subplots(n_factors, n_factors, figsize=(size*n_factors, size*n_factors))
#     fig.subplots_adjust(hspace=0.2, wspace=0.2)

    for i in range(n_factors):
        for j in range(n_factors):
            curr_ax = ax[i,j]
            if i == j:
                curr_ax.set_xlim([0,1])
                curr_ax.set_ylim([0,1])
                curr_ax.set_axis_off()
                curr_ax.text(0.5, 0.5, factors[i], fontsize=26, ha='center', va='center')
            else:
                r = np.round(data[factors[i]].corr(data[factors[j]]), 2)
                label = "r = %.2f" % r
                curr_ax = sns.regplot(data[factors[j]], data[factors[i]], ax=curr_ax, x_jitter=0.2, y_jitter=0.2, order=1, 
                            scatter_kws={'alpha': 0.2, 's': 26}, label=label, color='RoyalBlue')
                curr_ax.text(0.65, 0.1, label, fontsize=14, transform=curr_ax.transAxes)
            curr_ax.set_xticklabels("", fontsize=4)
            curr_ax.set_yticklabels("", fontsize=4)
            curr_ax.set_xlabel('')
            curr_ax.set_ylabel('')
#             curr_ax.xaxis.set_visible(False)
#             curr_ax.yaxis.set_visible(False)
    plt.tight_layout()
```

### Prediction

* Many relationships are seemingly strong

* Everything is highly significant

* How well can we actually predict Sexuality IAT bias without overfitting?

```python
""" Prepare the data for predictive modeling """

def get_Xy(data, y='IAT_overall', dummies=None, cont=None, recode_y=False):
    """ Select the initial variables to include in the model--
     categorical and continuous """
    
    if dummies is None:  # pick a reasonable default set
        dummies = ['sex', 'orientation', 'race', 'marriagelegal', 'contactdaily', 'contactfamily', 
              'contactfriend', 'contactregular', 'contactmet', 'major', 'lifestyle',
              'marriagerights', 'relationslegal', 'parent1', 'edu', 'STATE', ]
        
    if cont is None:
        cont = ['age', 'politicalid', 'religionid', 'tgaymen', 'tgaywomen', 'tstraightwomen', 
                    'tstraightmen']

    # Make sure feature set doesn't include outcome
    dummies = list(set(dummies) - set([y]))
    cont = list(set(cont) - set([y]))
    
    # Dummy-code all categorical variables
    dummies = [pd.get_dummies(data[f], prefix=f) for f in dummies]

    # Mean imputation for continuous variables
    if cont:
        def impute(x):
            x[np.isnan(x)] = x.mean()
            return x
        cont_data = data[cont].apply(impute)
        dummies.append(cont_data)

    # Merge all predictors
    full_X = pd.concat(dummies, axis=1)

    # List our final predictors
    names = full_X.columns # save for later

    full_X = full_X.values
    full_y = data[y]
    
    # Recode categorical outcome as numeric
    if recode_y:
        levels = list(full_y.unique())
        full_y = full_y.replace(dict(zip(levels, range(len(levels)))))
        
    full_y = full_y.values

    # Remove rows with NaN in y
    present = ~np.isnan(full_y)
    full_X = full_X[present]  # drop rows
    full_y = full_y[present]
        
    return (pd.DataFrame(full_X, columns=names), full_y)
```

```python
""" Generate a pipeline in scikit-learn and fit to IAT data """

from sklearn import preprocessing
from sklearn.pipeline import Pipeline
from sklearn import linear_model, svm, ensemble
from sklearn import cross_validation
from sklearn import feature_selection

def run_pipeline(X, y, n=10000, plot=True):
    """ Create a scikit-learn Pipeline and fit it to the data. """

    names = list(X.columns)
    X = X.values
    
    # Take a random subset of the dataset to cut down time
    if n is not None and n < len(X):
        rows = np.random.choice(np.arange(len(X)), n)
        X = X[rows]
        y = y[rows]

    # Set up a pipeline
    steps = []

    # Add imputation
    imputer = preprocessing.Imputer()
    steps.append(('impute', imputer))

    # Add scaling
    scaler = preprocessing.StandardScaler()
    steps.append(('scale', scaler))

    # Add feature selection
    selector = feature_selection.SelectKBest(k='all')
    steps.append(('select', selector))

    # Add estimator
    estimator = linear_model.Lasso(alpha=0.01)
#     estimator = svm.SVC(class_weight='auto')
    steps.append(('estimate', estimator))

    # Initialize the pipeline
    pipeline = Pipeline(steps)

    # Placeholder for predicted values
    y_predicted = np.zeros_like(y)

    i = 0

    # Set up cross-validation: training and test sets
    for train, test in cross_validation.KFold(len(X), n_folds=5, shuffle=True):

        i += 1

        # Split data into training and testing folds
        X_train, X_test, y_train, y_test = X[train], X[test], y[train], y[test]

        # Fit entire pipeline to training data
        pipeline.fit(X_train, y_train)

        # Apply fitted model to test data
        y_predicted[test] = pipeline.predict(X_test)

    # Fit to all of the data and save coefficients
    pipeline.fit(X, y)
    coefs = np.round(pipeline.steps[-1][1].coef_, 3)

    print "Cross-validated multiple R for model: %.2f\n" % np.corrcoef(y_predicted, y)[0,1]
    
    if plot:
        sns.jointplot(y_predicted, y, joint_kws={'alpha': 0.5})

    # print coefficients by rank
    coefs = pd.Series(coefs, index=names)
    print coefs.order('coef', ascending=False)
    
```

```python
X, y = get_Xy(data)
run_pipeline(X, y, n=10000, plot=True)
```
